# Bar plot
```sh
# bar
plt.bar(country, population)
plt.show()
# Styles
plt.bar(country, population, width=0.5, color="red")
plt.bar(country, population, width=0.5, color=["red", 'blue','green'])
# Styles axis x
plt.bar(country, population, width=0.5)
plt.xticks(np.arange(5), [x.title() for x in country], rotation=45) # Configuraciones de x
# bar horizon
plt.barh(country, population)
```

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
country = ['india', 'japan','mexico', 'colombia','germany']
population = [100,800,900,1000,300]
````

````python
plt.bar(country, population)
plt.show()
````

````python
plt.bar(country, population, width=0.5, color="red")
plt.show()
````

````python
plt.bar(country, population, width=0.5, color=["red", 'blue','green'])
plt.show()
````

````python
plt.bar(country, population, width=0.5)
# Configuraciones de x
plt.xticks(np.arange(5), [x.title() for x in country], rotation=45)
plt.show()
````

````python
plt.barh(country, population)
plt.show()
````

````python
profession = ['electronica', 'electricidad','informatica','mecanica','telecomunicaciones']
salary =  [10,20,30,10,20]
plt.barh(profession, salary)
plt.yticks(np.arange(5), [p.title() for p in profession], rotation=-5)
plt.show()
````
