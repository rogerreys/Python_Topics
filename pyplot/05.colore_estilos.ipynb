## Uso de Pyplot
```sh
# alpha -> transparencia
# linewidth  -> grosor
# linestyle -> stilo linea
# marker -> marcadores
# markersize | markerfacecolor

axs.plot(x,x+2, color='gray', alpha=0.5)
axs.plot(x,x+1, color='#ff33ff', linewidth = 10)
axs.plot(x,x+3, color='#333', linestyle=':')
axs.plot(x,x+4, color='#ff0055', linestyle='dashed')
axs.plot(x,x+5, linewidth=0.4, marker='p')
axs.plot(x,x+6, linewidth=0.4, marker='v', markersize=10, markerfacecolor='red')
```

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(0,6,11)
````

````python
print(plt.style.available)
````

````python
plt.style.use('grayscale')
````

````python
fig,ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+2, 'b')
ax.plot(x,x+1, 'r')
ax.plot(x,x+3, 'g')
````

````python
# Modificaciones de tipo matlap
fig, axes = plt.subplots(figsize=(6,6))
axes.plot(x,x+2, 'b--')
axes.plot(x,x+1, 'r-')
axes.plot(x,x+3, 'g.')
axes.plot(x,x-2, 'y.-')
axes.plot(x,x+4, 'ro:')
````

# Pyplot

````python
fig, axs = plt.subplots(figsize=(6,6))
# alpha -> transparencia
# linewidth  -> grosor
# linestyle -> stilo linea
# marker -> marcadores
# markersize | markerfacecolor

axs.plot(x,x+2, color='gray', alpha=0.5)
axs.plot(x,x+1, color='#ff33ff', linewidth = 10)
axs.plot(x,x+3, color='#333', linestyle=':')
axs.plot(x,x+4, color='#ff0055', linestyle='dashed')
axs.plot(x,x+5, linewidth=0.4, marker='p')
axs.plot(x,x+6, linewidth=0.4, marker='v', markersize=10, markerfacecolor='red')
````
