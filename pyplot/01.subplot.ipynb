# Subplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,8, 11)
y = x**2
print('x: ', x, '\ny: ',y)
```

```python
# plt.subplot(fila, columna, indice)
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.plot(y,x,'rv:')
plt.subplot(2,2,2)
plt.pie(x)
plt.subplot(2,2,3)
plt.plot(x,y, 'y>--')
plt.subplot(2,2,4)
plt.scatter(x,y)
plt.show()
```
