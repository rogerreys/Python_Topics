````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(0, 6, 11)
y = np.sin(x)
````

````python
fig, axes = plt.subplots()
axes.plot(x,y, 'b--')
````

````python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'g')
````

````python
fig, (axe1, axe2) = plt.subplots(nrows=1, ncols=2)
axe1.plot(x,y,'b')
axe2.plot(y,x,'g')
````

````python
fig, axe = plt.subplots(nrows=2, ncols=4)
axe[0,0].plot(x, np.sin(x))
axe[0,1].plot(x, np.cos(x), 'r')
axe[0,2].plot(x, np.tan(x), 'y')
axe[0,3].plot(x, np.sin(x)**2)
````

````python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(nrows=2, ncols=4)
ax1.plot(x,y, 'r^:')
ax2.plot(x,np.sin(y), 'r.')
ax3.hist(x)
ax4.pie(x)

ax5.scatter(x, y)
ax6.boxplot(y)
ax7.plot(x,np.tan(x))
ax8.pie(x**2)

fig.tight_layout()
````
