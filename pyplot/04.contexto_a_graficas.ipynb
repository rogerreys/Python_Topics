````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(0,5,11)
y = np.sin(x)
````

````python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,2))
#axes[0].plot(x, y, label="sin(x)") # Para aplique label, usar la funcion legend()
axes[0].plot(x, y, label="$sin(x)$") # Para aplique label, usar la funcion legend()
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y, x, label='sin(x)')
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
axes[1].legend()
fig.align_labels()
````

````python
plt.plot(x,y, label='sin(x)')
plt.title('Grafica sin subplots')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower left', bbox_to_anchor=(1, 0.5))
````
