# Grafica orientado a objetos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,6, 10)
y = x**2
```

```python
fig = plt.figure()
# [posicion,posicion,tamaño,tamaño]
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
fig.show()
```

```python
# Objetos distintos en una misma figura
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])
axes2.set_facecolor('gray')

axes.plot(x,y,'r--')
axes2.plot(y,x,'g:')
fig.show()
```
