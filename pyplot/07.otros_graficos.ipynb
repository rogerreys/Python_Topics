## Histogramas
```sh
# bins -> # de ramas
plt.hist(data, bins=20)
plt.hist(data, width=10)
# paso trasparente
plt.hist(data, bins=20, histtype='step')
```
## Boxplot
```sh
plt.boxplot(data)
# Cambio de direccion
plt.boxplot(data, vert=False)
# Colorea box
plt.boxplot(data, vert=False, patch_artist=True)
# Muestra la mediana
plt.boxplot(data, vert=False, notch=True)
# outlayer fuera del rango
# showfliers - muestra o no los fuera de rango 
data = np.append(data, 200)
# plt.boxplot(data, vert=False, showfliers=False)
plt.boxplot(data, vert=False, showfliers=True)
```

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
data = np.random.randint(1,50,100)
````

````python
# bins -> # de ramas
plt.hist(data, bins=20)
plt.show()
````

````python
plt.hist(data, width=10)
plt.show()
````

````python
plt.hist(data, bins=20, histtype='step')
plt.show()
````

# Boxplot

````python
plt.boxplot(data)
plt.show()
````

````python
plt.boxplot(data, vert=False)
plt.show()
````

````python
plt.boxplot(data, vert=False, patch_artist=True)
plt.show()
````

````python
# Muestra la mediana
plt.boxplot(data, vert=False, notch=True)
plt.show()
````

````python
# outlayer fuera del rango
# showfliers - muestra o no los fuera de rango 
data = np.append(data, 200)
# plt.boxplot(data, vert=False, showfliers=False)
plt.boxplot(data, vert=False, showfliers=True)
plt.show()
````
