````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
tips = sns.load_dataset('tips')
tips.head(3)
````

````python
#sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.6))
plt.show()
````

````python
marker = {"Lunch":"D", "Dinner":"s"}
plt.figure(figsize=(6,6))
#sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=marker)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.6))
plt.show()
````

````python
sns.lineplot(data=tips, x='total_bill', y='tip',hue='day', style='time',size='size')
plt.legend(loc='center', bbox_to_anchor=(1.15,0.5))
plt.show()
````

````python
# sns.relplot(data=tips, x='total_bill', y='tip', hue='day', size='size', style='time')
#sns.relplot(data=tips, x='total_bill', y='tip', hue='day', size='size', style='time', kind='scatter')
#sns.relplot(data=tips, x='total_bill', y='tip', hue='day', size='size', style='time', kind='line')
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', size='size', style='time', kind='scatter', col='time')
plt.show()
````
