````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
tips = sns.load_dataset('tips')
tips.head(5)
````

````python
#sns.countplot(data=tips, x='day', hue='sex')
sns.countplot(data=tips, y='day', hue='sex')
plt.show()
````

````python
#sns.stripplot(data=tips, x='day', hue='sex')
#sns.stripplot(data=tips, x='day', y='total_bill', hue='sex')
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
````

````python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
````

````python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
````

````python
plt.figure(figsize=(8,8))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', color="0", marker='v')
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
````

````python
plt.figure(figsize=(6,6))
#sns.violinplot(data=tips, x='day', y='total_bill', hue='sex')
#sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, palette='pastel')
sns.violinplot(data=tips, y='total_bill')
plt.show()
````

````python
plt.figure(figsize=(6,6))
#sns.catplot(data=tips, x='day', y='total_bill', hue='sex')
#sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='box')
#sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='swarm')
#sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='violin')
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='violin', col='time')
plt.show()
````
