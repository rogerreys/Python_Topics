````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
tips = sns.load_dataset('tips')
tips
````

````python
sns.displot(data=tips, x='total_bill')
plt.show()
````

````python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
````

````python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
````

````python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde') # distribucion por densidad kde
````

````python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=False, palette='dark', alpha=0.25)
````
