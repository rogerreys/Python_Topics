````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
tps = sns.load_dataset('tips')
tps.head(5)
````

````python
sns.histplot(data=tps, x='tip', bins=50)
plt.show()
````

````python
# Mire el comportamiento acumulativos
sns.histplot(data=tps, x='tip', bins=50, cumulative=True)
plt.show()
````

````python
sns.histplot(data=tps, x='tip', bins=50, hue='sex', stat='frequency')
````

````python
sns.histplot(data=tps, x='tip', bins=50, hue='sex', stat='probability')
````

````python
sns.histplot(data=tps, x='tip', bins=50, hue='sex', stat='percent')
plt.show()
````

````python
sns.histplot(data=tps, x='tip', bins=30, stat='density')
plt.show()
````

````python
# sns.histplot(data=tps, x='tip',hue='sex', bins=30, stat='count', multiple='stack')
# sns.histplot(data=tps, x='tip',hue='sex', bins=30, stat='count', multiple='layer')
# sns.histplot(data=tps, x='tip',hue='sex', bins=30, stat='count', multiple='dodge')
sns.histplot(data=tps, x='tip',hue='sex', bins=30, stat='count', multiple='fill')
````

````python
#sns.kdeplot(data=tps, x='tip', hue='sex')
#sns.kdeplot(data=tps, x='tip',hue='sex', cumulative=False, shade=True)
sns.kdeplot(data=tps, x='tip',hue='sex', cumulative=False, shade=True, bw_adjust=1)
plt.show()
#tps.head()
````

````python
#sns.ecdfplot(data=tps, x='tip', hue='sex')
sns.ecdfplot(data=tps, x='tip', hue='sex',stat='count')
plt.show()
````

````python
#sns.displot(data=tps, x='tip', hue='sex')
#sns.displot(data=tps, x='tip', hue='sex', kind='kde')
#sns.displot(data=tps, x='tip', hue='sex', kind='hist')
sns.displot(data=tps, x='tip', hue='sex', kind='hist', multiple='stack')
plt.show()
````
