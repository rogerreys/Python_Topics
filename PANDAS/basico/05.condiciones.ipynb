# Condiciones
```sh 
# Mayor que
df[df['Year']>2016]
```
```sh
# Igual igual
df[df['Genre']=='Fiction']
```
```sh
# And
df[(df['Year']>2018) & (df['Genre']=='Non Fiction')]
```
```sh
# Negacion
df[-(df['Price']>10)]
```

````python
df[(df['Year']>2018) & (df['Genre']=='Non Fiction')]
````
