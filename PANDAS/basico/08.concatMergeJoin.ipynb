# Merge Join
- Left join
- Right join
- Inner join
- Outer join
- Concart
    Axis 0 - Por fila (Verticalmente)
    Axis 1 - Por columna (Horizontalmente)

```sh
# concat
pd.concat([df1, df2])
pd.concat([df1, df2], ignore_index=True)  # ignora secuencia de index
pd.concat([df1, df2], axis=1)   # concat columna
```
```sh
# Merge
izq.merge(der)
izq.merge(der, on='key')  # Especifica columna
izq.merge(der, left_on='key', right_on='key 2')

izq.merge(der, left_on='key', right_on='key 2', how='left')   # union izquierda
izq.merge(der, left_on='key', right_on='key 2', how='right')   # union derecha
```

````python
import pandas as pd
import numpy as np
````

````python
pd.concat([df1, df2], ignore_index=True)
````

## Merge

````python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
'A':['A0','A1','A2','A3'],
'B':['B0','B1','B2','B3']})
der = pd.DataFrame({'key':['k0','k1','k2','k3'],
'C':['C0','C1','C2','C3'],
'D':['D0','D1','D2','D3']})
der
````

````python
izq.merge(der, on='key')
````

````python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
'A':['A0','A1','A2','A3'],
'B':['B0','B1','B2','B3']})
der = pd.DataFrame({'key 2':['k0','k1','k2','k3'],
'C':['C0','C1','C2','C3'],
'D':['D0','D1','D2','D3']})

izq.merge(der, left_on='key', right_on='key 2')
````

````python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
'A':['A0','A1','A2','A3'],
'B':['B0','B1','B2','B3']})
der = pd.DataFrame({'key 2':['k0','k1','k2',np.nan],
'C':['C0','C1','C2','C3'],
'D':['D0','D1','D2','D3']})

print("Merge sin nan\n", izq.merge(der, left_on='key', right_on='key 2'))
print("Merge solo de izquierda\n", izq.merge(der, left_on='key', right_on='key 2', how='left'))
print("Merge solo de derecha\n", izq.merge(der, left_on='key', right_on='key 2', how='right'))
````
