# Groupby
```sh
# Funciones con groupby
df.groupby('Author').count()

df.groupby('Author').mean()

df.groupby('Author').sum()

df.groupby('Author').min()

df.groupby('Author').max()

df.groupby('Author').sum().loc['Stephen King']

# Resetear groupby
df.groupby('Author').sum().reset_index()


# Uso de varios funciones
df.groupby('Author').agg(['min','max'])

df.groupby('Author').agg({'Reviews':['min', 'max'], 'User Rating':['sum']})


# Llave compuesta
df.groupby(['Author','Year']).count()
```

````python
import pandas as pd
````

````python
df = pd.read_csv('./assts/bestsellers-with-categories.csv')
df.head(2)
````

````python
df.groupby('Author').count()
````

````python
print('Mean\n',df.groupby('Author').mean())
print('Sum\n',df.groupby('Author').sum())
````

````python
df.groupby('Author').min()
````

````python
df.groupby('Author').max()
````

````python
df.groupby('Author').sum().loc['Stephen King']
````

````python
# Resetear groupby
df.groupby('Author').sum().reset_index()
````

````python
df.groupby('Author').agg(['min','max'])
````

````python
df.groupby('Author').agg({'Reviews':['min', 'max'], 'User Rating':['sum']})
````

````python
# Llave compuesta
df.groupby(['Author','Year']).count()
````
